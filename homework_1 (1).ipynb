2. Write a Python program that:

a) Asks the user for their name. (hint: use input(), in case of any trouble just google it!)

b) Greets the user with their name and tells them how many letters are in their name.

c) Converts their name to uppercase and prints it.

```python
name = str(input("Enter you name: "))
print(f"Hello {name}, in your name there are {len(name)} letters")
print(name.upper())
```

3. Write a Python program that:

Converts temperatures from Celsius to Fahrenheit.

The formula for conversion is: Fahrenheit = (Celsius * 9/5) + 32

Ask the user for the temperature in Celsius (hint: again using input()) and print the equivalent Fahrenheit temperature.

```python
celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = (celsius * 9/5) + 32
print("Temperature in Fahrenheit:", fahrenheit)
```

3. Write a Python program that:

a) Asks the user for their first and last name.

b) Concatenates the first and last name with a space in between.

c) Prints the full name and its length.

d) Make manipulations with indices of full_name string

```python
first_name = str(input("Enter your first name: "))
last_name = str(input("Enter your second name:"))
full_name = first_name + " " + last_name
print(full_name)
print(len(full_name))
print("First character:", full_name[0])
print("Last character:", full_name[-1])
print("First 3 characters:", full_name[:3])
```
